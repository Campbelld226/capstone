# Visualizations

```python
import pandas as pd
from Labeler import Labeler
import matplotlib.pyplot as plt 
import warnings
import seaborn as sns
import random
pd.options.display.max_rows = 999
warnings.filterwarnings('ignore')
```

# Clustering Algorithms

## HAT-P-23b

```python
# Agglomerative
l1 = Labeler(pd.read_csv('test/HAT-P-23b_1.2.csv'))
l1.get_data()
agglom = l1.cluster_method(1)
l1.order_clusters(agglom)

# Spectral
l2 = Labeler(pd.read_csv('test/HAT-P-23b_1.2.csv'))
l2.get_data()
spectral = l2.cluster_method(2)
l2.order_clusters(spectral)

# K-Means
l3 = Labeler(pd.read_csv('test/HAT-P-23b_1.2.csv'))
l3.get_data()
kmeans = l3.cluster_method(3)
l3.order_clusters(kmeans)

# Graph clustering Methods
fig, ax = plt.subplots(1,3, figsize=(15,5), tight_layout=1)
ax[0].set(title="Agglomerative", xlabel="Epoch")
ax[1].set(title="Spectral", xlabel="Epoch")
ax[2].set(title="K-Means", xlabel="Epoch")
sns.scatterplot(x=range(len(l1.X['depth'])),
                    y=l1.X['depth'] * -1, hue=l1.X['labels_desc'], ax=ax[0])
sns.scatterplot(x=range(len(l2.X['depth'])),
                    y=l2.X['depth'] * -1, hue=l2.X['labels_desc'], ax=ax[1])
sns.scatterplot(x=range(len(l3.X['depth'])),
                    y=l3.X['depth'] * -1, hue=l3.X['labels_desc'], ax=ax[2])
fig.show()
```

## Stats on Each Attribute

```python
df = l1.X[[x for x in l1.X if x not in ['is_level', 'is_pos', 'is_neg', 'labels_int']]]
df.describe()
```

## Scatter and Distribution Plot of Each Attribute

```python
for index, col in enumerate(df):
    if df[col].dtype != (object):
        fig, ax = plt.subplots(1,2, figsize=(12, 5), tight_layout=1)
        plt.title(col)
        sns.scatterplot(x=range(len(df[col])), y=df[col], ax=ax[0])
        sns.distplot(df[col], ax=ax[1])
        fig.show()
```
