# Visualizations

```python
import pandas as pd
from Labeler import Labeler
import matplotlib.pyplot as plt 
import warnings
import seaborn as sns
pd.options.display.max_rows = 999
# warnings.filterwarnings('ignore')
```

## Clustering Algorithms

```python
l = Labeler(pd.read_csv('test/no_transit.csv'))
df = l.get_data()
df
```

## Countplot of unique values

```python
display(df['labels_desc'].value_counts())
fig = plt.figure(figsize=(8,5))
sns.countplot(x='labels_desc',data = df)
plt.title('Count of Unique Periods')
```

## Stats on Each Attribute

```python
df = df[[x for x in df if x not in ['is_level', 'is_pos', 'is_neg', 'labels_int']]]
df.describe()
```

## Scatter and Distribution Plot of Each Attribute

```python
for index, col in enumerate(df):
    if df[col].dtype != (object):
        fig, ax = plt.subplots(1,2, figsize=(12, 5), tight_layout=1)
        plt.title(col)
        sns.scatterplot(x=range(len(df[col])), y=df[col], ax=ax[0])
        sns.distplot(df[col], ax=ax[1])
        fig.show()
```
